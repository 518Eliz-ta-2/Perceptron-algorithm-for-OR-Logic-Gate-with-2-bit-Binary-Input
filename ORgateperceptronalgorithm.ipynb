```python
import numpy as np
```

Step activation function

```python
def unit_step(a):
    if a >= 0:
       return 1
    else:
       return 0
```

single perceptron neuron model

```python
def perceptron_model(x,w,b):
    a = np.dot(w,x) + b #weights*input + bias(threshold)
    y = unit_step(a)    #output
    return y
```

OR gate logic

```python
#OR LOGIC FUNCTION
# w1 = 2, w2 = 1, b = -0.5
def or_gate(x):
    w = np.array([2,1])
    b = -0.5
    return perceptron_model(x,w,b)

tests = [(0,1), (1,1), (0,0), (1,0)]

for x in tests:
    print(f"OR{x} = {or_gate(np.array(x))}")
```
